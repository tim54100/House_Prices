```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#讀入資料
df_train = pd.read_csv('./data/train.csv')
```

```python
#檢查資料
df_train.columns
```

### 看起來資料有許多細項，那我們該如何選擇要餵入的資料呢?
- 首先SalePrice看起來是我們想預測的
- 那我們可以來找找哪些資料和SalePrice有關聯
- 再加以篩選，看看哪些資料之間太有關聯，選擇留下和SalePrice較高關聯的資料

```python
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
corr_sale = (corrmat[corrmat['SalePrice'] > 0.5])['SalePrice'] #找出關聯性大於0.5的
corr_sale.drop('SalePrice',inplace=True) # 
print(corr_sale.index)
```

```python
#a = [] 用來看看是哪些值有關連的list
de = []
for i in corr_sale.index:
    for j in corr_sale.index:
        if corrmat[i][j] > 0.7 and i!=j and (not i in de) and (not j in de):
            if corr_sale[i] > corr_sale[j]:
                corr_sale.drop(j, inplace=True)
                #a += [[i,j]]
                de += [j]
            else:
                corr_sale.drop(i, inplace=True)
                #a += [[i,j]]
                de += [i]

print(de) #印出被剃除的資料
print(corr_sale) #印出剩下的資料
```

### 資料有可能遺漏
所以我們應該試著看看之前篩選出來的資料是否完整

```python
miss = {}
for i in corr_sale.index: #藉由之前篩選出來的資料，來看看他們有多少值遺漏
    #if df_train[i].isnull().sum()>0:
    miss.update({i:df_train[i].isnull().sum()})
print(miss) #印出後發現都沒有遺漏，代表不需要處理
```

 ### 離群值
 - 有些資料中的值可能比較太小或太大，有可能會是輸入錯誤，或其他原因
 - 而他們會影響到我們的預測，所以必須把其剃除掉
 - 在這裡，我們簡單用上四分位數和下四分位數來預測離群估測值

```python
for i in corr_sale.index:
    data = pd.concat([df_train['SalePrice'], df_train[i]], axis=1)
    data.plot.scatter(x=i, y='SalePrice', title="before", ylim=(0,800000));
    q1, medium, q3 = np.percentile(df_train[i], [25, 50, 75])
    k = 3
    h_e, l_e = q3 + k*(q3-q1), q1 - k*(q3-q1)
    #print(df_train[i])
    high = (df_train[df_train[i] > h_e])[i]
    #print(high.index)
    df_train.drop(high.index, inplace=True)
    low = (df_train[df_train[i] < l_e])[i]
    #print(low.index)
    df_train.drop(low.index, inplace=True)
    data = pd.concat([df_train['SalePrice'], df_train[i]], axis=1)
    data.plot.scatter(x=i, y='SalePrice', title="after", ylim=(0,800000));
```

### 正規化
簡單而言，我們要讓數據看起來是常態分布，有許多統計的測試都依賴這個。
<br>
我們用兩種圖來觀測資料是否為常態分佈
- Histogram - 以藍色線表示目前的分布情形，黑色線為理想常態分佈
- Normal probability plot- 以藍色點表示資料分布情況，以紅線表示理想常態分佈

```python
#用SalePrice畫出這兩種圖
sns.distplot(df_train['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

我們看得出剛剛的圖形明顯表示其並不是常態分佈，在統計學上呈現positive skewness，即Histogram的峰偏左。<br>
我們可以簡單用log來處理，通常用log來轉換後成果都很不錯。

```python
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

```python
sns.distplot(df_train['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

可以看到他的資料已經很接近常態分佈了<br>
接下來我們對一樣是positive skewness做一樣的事，不過得注意等於0的數值，他們不能被log

```python
remaining = []
for i in corr_sale.index:
    if (df_train[i]==0).sum() == 0 and df_train[i].skew()>0:
        #res = stats.probplot(df_train[i], plot=plt)
        #df_train['SalePrice'] = np.log(df_train['SalePrice'])
        #sns.distplot(df_train[i], fit=norm);
        df_train[i] = np.log(df_train[i])
        sns.distplot(df_train[i], fit=norm);
        fig = plt.figure()
        res = stats.probplot(df_train[i], plot=plt)
    else:
        remaining += [i]
print("remaining :" + str(remaining))
```

```python
#來看看剩餘的資料中分別有甚麼狀況
for i in remaining: 
    print(i + '\nis0: '+ str((df_train[i] == 0).sum()) + '\n skewness: ' + str(df_train[i].skew()))
```
